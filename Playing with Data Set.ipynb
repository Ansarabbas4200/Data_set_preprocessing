## Minimum and Maximum Salary

```python
import pandas as pd
```

```python
data = pd.read_csv("D:\ProgrammerForceClass\Pandas\ds_salaries.csv")
data
```

```python
data_analyst = data[data.job_title == "Data Analyst"]
data_analyst
```

```python
data_analyst = data_analyst[['salary_in_usd', 'company_size']]
data_analyst
```

```python
data_analyst_max = data_analyst[data_analyst.salary_in_usd == data_analyst.salary_in_usd.max()]
data_analyst_max
```

```python
data_analyst_min = data_analyst[data_analyst.salary_in_usd == data_analyst.salary_in_usd.min()]
data_analyst_min
```

#### Maximum and minimun salaries with their company size

```python
final_data = pd.concat([data_analyst_max, data_analyst_min])
final_data
```

```python
final_data.to_csv("D:\ProgrammerForceClass\Pandas\max_min_salary.csv")
```

## Region as per have more jobs data engineer

```python
data_engineers = data[data.job_title.str.contains('data engineer', case=False)]
data_engineers
```

```python
jobs_by_location = data_engineers.groupby('company_location')['job_title'].count().reset_index()
jobs_by_location
```

```python
sorted_jobs_by_location = jobs_by_location.sort_values('job_title', ascending = False)
sorted_jobs_by_location.to_csv("D:\ProgrammerForceClass\Pandas\most_jobs_region.csv")
sorted_jobs_by_location

```
